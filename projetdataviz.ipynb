```python
# 1) Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import timedelta

# set plotting defaults
%matplotlib inline
plt.rcParams['figure.figsize'] = (10,5)

# 2) Chargement
DATA_PATH = DATA_PATH = '/Users/hajarblmq/Downloads/online_retail_II.xlsx'
  # modifier si nécessaire
try:
    df = pd.read_csv(DATA_PATH, encoding='ISO-8859-1', parse_dates=['InvoiceDate'])
except Exception:
    df = pd.read_excel(DATA_PATH, sheet_name=0, parse_dates=['InvoiceDate'])

# 3) Fiche synthétique
def fiche_synthese(df):
    synth = {
        'rows': len(df),
        'cols': len(df.columns),
        'period_min': df['InvoiceDate'].min(),
        'period_max': df['InvoiceDate'].max(),
        'n_customers': df['Customer ID'].nunique(dropna=True)
    }
    return synth

print('Fiche synthèse :')
print(fiche_synthese(df))

```
